```python
# how far back does the data go?
STORM_ID = '1611661328'

def check_season(season):
    try:
        df = leaguedashlineups.LeagueDashLineups(
            season=season,
            season_type_all_star='Regular Season',
            league_id_nullable='10',
            measure_type_detailed_defense='Base',
            per_mode_detailed='Per100Possessions',
            team_id_nullable=STORM_ID
        ).get_data_frames()[0]
        print(f"{season}: {len(df)} lineups returned")
        time.sleep(5)
    except Exception as e:
        print(f"{season}: FAILED — {e}")

# Spot check a spread of years
for season in ['2024', '2020']:
    check_season(season)
```

```python
# 2021 might be first year. investigate

def check_season_verbose(season):
    try:
        endpoint = leaguedashlineups.LeagueDashLineups(
            season=season,
            season_type_all_star='Regular Season',
            league_id_nullable='10',
            measure_type_detailed_defense='Base',
            per_mode_detailed='Per100Possessions',
            team_id_nullable=STORM_ID
        )
        df = endpoint.get_data_frames()[0]
        print(f"{season}: {len(df)} lineups returned")
        time.sleep(1)
    except Exception as e:
        print(f"{season}: FAILED — {e}")

# Binary search approach — start wide then narrow
for season in ['1997', '1998', '1999', '2000', '2001', '2020']:
    check_season_verbose(season)

# data starts in 2000
```

```python
# the Storm began play in 2000, so that checks out
# lets pull all Storm lineup data history

STORM_ID = '1611661328'

def get_storm_lineups(season, measure_type='Base'):
    try:
        df = leaguedashlineups.LeagueDashLineups(
            season=season,
            season_type_all_star='Regular Season',
            league_id_nullable='10',
            measure_type_detailed_defense=measure_type,
            per_mode_detailed='Per100Possessions',
            team_id_nullable=STORM_ID
        ).get_data_frames()[0]
        
        if len(df) == 0:
            print(f"  {season} {measure_type}: empty")
            return None
            
        df['SEASON'] = season
        df['MEASURE_TYPE'] = measure_type
        print(f"  {season} {measure_type}: {len(df)} lineups")
        time.sleep(10)  # conservative sleep
        return df
        
    except Exception as e:
        print(f"  {season} {measure_type}: FAILED — {e}")
        time.sleep(30)  # longer sleep on failure
        return None

seasons = [str(y) for y in range(2000, 2025)]

base_frames = []
adv_frames = []

for season in seasons:
    print(f"Fetching {season}...")
    
    base = get_storm_lineups(season, 'Base')
    if base is not None:
        base_frames.append(base)
    
    adv = get_storm_lineups(season, 'Advanced')
    if adv is not None:
        adv_frames.append(adv)

# Only concat non-empty results
if base_frames:
    base_df = pd.concat(base_frames, ignore_index=True)
    base_df.to_csv('../data/raw/storm_lineups_base.csv', index=False)
    print(f"\nBase shape: {base_df.shape}")
    print(f"Seasons with data: {sorted(base_df['SEASON'].unique())}")

if adv_frames:
    adv_df = pd.concat(adv_frames, ignore_index=True)
    adv_df.to_csv('../data/raw/storm_lineups_advanced.csv', index=False)
    print(f"Advanced shape: {adv_df.shape}")
```
